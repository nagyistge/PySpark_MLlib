```python
# load census population data
# sex ratio = how many males per 100 females
df_pop = pd.read_csv('census_pop_data.txt', sep = '\t')
df_pop.head(52)
```

```python
df_pop['Under 18 years Percent'] = df_pop['Under 18 years Percent'].apply(lambda x: x/100)
df_pop['18 to 44 years Percent'] = df_pop['18 to 44 years Percent'].apply(lambda x: x/100)
df_pop['45 to 64 years Percent'] = df_pop['45 to 64 years Percent'].apply(lambda x: x/100)
df_pop['65 years and over Percent'] = df_pop['65 years and over Percent'].apply(lambda x: x/100)
```
